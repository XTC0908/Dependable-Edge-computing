```python
import osmnx as ox
import networkx as nx
%matplotlib inline
```

```python
location_point = (59.34, 18.07)
G = ox.graph_from_point(location_point, distance=500, simplify=False, network_type='drive')
G = ox.project_graph(G)
#G = ox.simplify_graph(G)
fig, ax = ox.plot_graph(G)
```

```python
nl = list(G.node)
nc = ['r' if (i == 1 or i == 72) else 'b' for i in range(len(nl))]
```

```python
fig, ax = ox.plot_graph(G, node_color=nc)
```
